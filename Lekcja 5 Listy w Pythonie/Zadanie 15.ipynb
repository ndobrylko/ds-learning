```python
logs = [
    "2024-01-15 10:23:45 INFO Model training started",
    "2024-01-15 10:45:12 WARNING Low memory",
    "2024-01-15 11:02:33 ERROR Training failed",
    "2024-01-15 11:05:01 INFO Retrying with smaller batch"
]
log_counts = {}

for log in logs:
    parts = log.split()
    level = parts[2]

    if level not in log_counts:
        log_counts[level] = 0

    log_counts[level] += 1

total_logs = len(logs)
log_percentages = {}

for level, count in log_counts.items():
    percentage = (count / total_logs) * 100
    log_percentages[level] = percentage

print(f"{'Poziom':<10}{'Liczba':>8}{'Procent':>12}")
print("-" * 30)

for level in log_counts:
    print(
        f"{level:<10}"
        f"{log_counts[level]:>8}"
        f"{log_percentages[level]:>11.1f}%"
    )
```
