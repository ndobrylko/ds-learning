```python
studenci = [
    ('Anna', 'Kowalska', 5.0, 4.5, 5.0),
    ('Jan', 'Nowak', 3.0, 3.5, 3.0),
    ('Maria', 'Zielińska', 4.0, 4.5, 5.0),
    ('Piotr', 'Wiśniewski', 2.0, 3.0, 2.5),
    ('Ewa', 'Wójcik', 5.0, 5.0, 4.5),
    ('Tomasz', 'Kamiński', 4.0, 3.5, 4.0)
]

#Słownik z 3 kategoriami
wynik = {
    'wyróżniający': [],
    'dobry': [],
    'średni': []
}

for student in studenci:
    imie, nazwisko, o1, o2, o3 = student
    srednia = (o1 + o2 + o3) / 3

    student_info = (imie, nazwisko, round(srednia, 2))

    if srednia >= 4.5:
        wynik['wyróżniający'].append(student_info)
    elif srednia >= 3.5:
        wynik['dobry'].append(student_info)
    else:
        wynik['średni'].append(student_info)

for kategoria, lista_studentow in wynik.items():
    print(f"\n{kategoria.upper()}:")
    for s in lista_studentow:
        print(s)
```
