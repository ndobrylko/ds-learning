```python
dzien1 = {'user001', 'user002', 'user003', 'user004', 'user005', 'user010'}
dzien2 = {'user002', 'user003', 'user006', 'user007', 'user008', 'user010'}
dzien3 = {'user003', 'user004', 'user009', 'user010', 'user011'}


użytkownicy_aktywni_3dni = dzien1 & dzien2 & dzien3

tylko_dzien1 = dzien1 - dzien2 - dzien3
tylko_dzien2 = dzien2 - dzien1 - dzien3
tylko_dzien3 = dzien3 - dzien1 - dzien2

uzytkownicy_aktywni_2dni = (
    (dzien1 & dzien2) |
    (dzien1 & dzien3) |
    (dzien2 & dzien3)
) - użytkownicy_aktywni_3dni

print("Użytkownicy aktywni wszystkie 3 dni:")
print(użytkownicy_aktywni_3dni)

print("\nUżytkownicy tylko 1 dzień:")
print("Dzień 1:", tylko_dzien1)
print("Dzień 2:", tylko_dzien2)
print("Dzień 3:", tylko_dzien3)

print("\nUżytkownicy aktywni dokładnie 2 dni:")
print(uzytkownicy_aktywni_2dni)
```
