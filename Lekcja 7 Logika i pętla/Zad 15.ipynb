```python
results = [78, 82, 85, 88, 90, 92, 95, 97, 99, 45, 101, 103, 105, 180]

results_sorted = sorted(results)
print(results_sorted)
```

```python
lower_half = results_sorted[:7]
upper_half = results_sorted[7:]

print("Dolna połowa:", lower_half)
print("Górna połowa:", upper_half)
```

```python
Q1 = lower_half[len(lower_half) // 2]
Q3 = upper_half[len(upper_half) // 2]
print("Q1:", Q1)
print("Q3:", Q3)
```

```python
IQR = Q3 - Q1
print("IQR:", IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Dolna granica:", lower_bound)
print("Górna granica:", upper_bound)
```

```python
outliers = []

for value in results_sorted:
    if value < lower_bound or value > upper_bound:
        outliers.append(value)

print("Outliers:", outliers)
```
